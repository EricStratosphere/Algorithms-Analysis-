```python
import matplotlib.pyplot as plt
import numpy as np

DFSTraversal = "DFSTime.txt"
BFSTraversal = "BFSTime.txt"
x = np.arange(1000, 11000, 1000)

with open(DFSTraversal, "r") as file:
    DFSTraversalArr = [float(line.strip()) for line in file]
with open(BFSTraversal, "r") as file:
    BFSTraversalArr = [float(line.strip()) for line in file]

plt.plot(x, DFSTraversalArr, label = "DFS Traversal Time", marker = 'o', color = "blue")
plt.plot(x, BFSTraversalArr, label = "BFS Traversal Time", marker = 'o', color = "green")

plt.title("DFS vs BFS: A comparison of execution times.")
plt.ylabel("Time")
plt.xlabel("Input size")
plt.legend(loc = "upper left", title = "Search Trees")
plt.grid(True)
plt.show()
```

# DFS vs BFS Execution Time Comparison

This document compares the execution time (in seconds) of **Depth-First Search (DFS)** and **Breadth-First Search (BFS)** algorithms over increasing dataset sizes or traversal depthscer.

## Analysis

- **Performance Comparison:**  
  DFS tends to execute faster than BFS in most test cases. For instance:
  - At Test Case 1, which has a starting dataset size of 1000, DFS is approximately **42% faster** than BFS.
  - By Test Case 10, which has a datase tsize of 10000, the difference narrows, with DFS only being about **2.7% faster** than BFS.

## Conclusion

- **DFS** is generally more time-efficient in this dataset, particularly for smaller graphs or lower traversal depths.
- **BFS**, while slightly slower, may offer benefits in certain scenarios.
- As input size increases, the difference in execution time between the two narrows, suggesting similar time complexity characteristics in large-scale graphs.

However, it is important to note that performance can vary depending on the graph structure and implementation.

## Implementation
- The way the process of extracting data is implemented in the code is as follows:
   -  An outer for-loop iterates i from 1000 to 10000 by an increment of 1000 per iteration, (1000, 2000, 3000, ..., 10000). In each iteration, a graph of size i is declared. An inner for-loop j traverses from 0 to i - 1. And in each iteration of j, another inner for-loop k traverses from j + 1 until i. The reason for this is for a randomizer function to execute a 50/50 chance of drawing a connection or adding an edge between vertex j and vertex k. If an edge is drawn between them, the source vertex is appended to a file. After the two inner loops are executed, a random source is extracted from the Sources.txt file. This is to ensure that any arbitrary vertex we choose for our starting vertex is guaranteed to have an edge. The selected starting vertex is then used as the starting point for the BFS and DFS functions to be tested for execution time.
