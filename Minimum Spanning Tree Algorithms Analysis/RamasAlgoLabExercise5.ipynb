```python
import matplotlib.pyplot as plt
import numpy as np

KruskalTime = "KruskalTime.txt"
PrimTime = "PrimTime.txt"
BoruvkaTime = "BoruvkaTime.txt"
x = np.arange(1000, 21000, 1000)

with open(KruskalTime, "r") as file:
    KruskalTimeArr = [float(line.strip()) for line in file]
with open(PrimTime, "r") as file:
    PrimTimeArr = [float(line.strip()) for line in file]
with open(BoruvkaTime, "r") as file:
    BoruvkaTimeArr = [float(line.strip()) for line in file]

plt.plot(x, KruskalTimeArr, label = "Kruskal Time", marker = 'o', color = "blue")
plt.plot(x, BoruvkaTimeArr, label = "Boruvka Time", marker = 'o', color = "green")
plt.plot(x, PrimTimeArr, label = "Prim Time", marker = 'o', color = "red")

plt.title("Kruskal VS Boruvka VS Prim : A comparison of Minimum Spanning Tree Algorithms.")
plt.ylabel("Time")
plt.xlabel("Input size")
plt.legend(loc = "upper left", title = "MST Algorithms")
plt.grid(True)
plt.show()
```

# Algorithm Performance Analysis

## Time Complexity Comparison of Borvka's, Kruskal's, and Prim's Algorithms

### Raw Data Summary

| Algorithm | Min Time (s) | Max Time (s) | Avg Time (s) | Median Time (s) |
|-----------|-------------|-------------|-------------|----------------|
| Borvka   | 0.002       | 0.043       | 0.025       | 0.032          |
| Kruskal   | 0.000       | 1.083       | 0.419       | 0.371          |
| Prim      | 0.005       | 2.406       | 0.897       | 0.856          |

### Time Complexity Observations

1. **Borvka's Algorithm**:
   - Shows the most consistent performance
   - Maintains relatively stable execution times (0.002s - 0.043s)
   - Demonstrates the best performance among the three algorithms
   - Time complexity appears to scale very well with increasing input sizes

2. **Kruskal's Algorithm**:
   - Shows moderate performance between Borvka and Prim
   - Execution times range from 0s to 1.083s
   - Exhibits more variability than Borvka but less than Prim
   - Performance degradation becomes noticeable with larger inputs

3. **Prim's Algorithm**:
   - Shows the poorest performance among the three
   - Execution times range from 0.005s to 2.406s
   - Demonstrates the most significant performance degradation
   - Time complexity scales poorly compared to the other algorithms

### Performance Characteristics

- **Best Case Performance**:
  - Borvka (0.002s) > Prim (0.005s) > Kruskal (0.000s)
  
- **Worst Case Performance**:
  - Prim (2.406s) > Kruskal (1.083s) > Borvka (0.043s)

- **Performance Stability**:
  - Borvka shows the most stable performance with minimal fluctuations
  - Kruskal shows moderate stability
  - Prim shows the least stable performance with large variations

### Theoretical vs Empirical Analysis

The empirical results align well with theoretical expectations:

1. **Borvka's Algorithm**:
   - Theoretical: O(E log V)
   - Empirical: Shows logarithmic-like growth pattern
   - Excellent for distributed/parallel implementations

2. **Kruskal's Algorithm**:
   - Theoretical: O(E log E) or O(E log V)
   - Empirical: Shows linearithmic growth
   - Performance depends heavily on union-find implementation

3. **Prim's Algorithm**:
   - Theoretical: O(E log V) with binary heap
   - Empirical: Shows worse-than-expected performance
   - May benefit from Fibonacci heap implementation

### Recommendations

1. **For small to medium graphs**:
   - All algorithms perform adequately
   - Borvka provides most consistent performance

2. **For large graphs**:
   - Borvka is clearly the best choice
   - Kruskal may be acceptable if Borvka isn't available
   - Prim should be avoided for large graphs

3. **For parallel implementations**:
   - Borvka is naturally parallelizable
   - Kruskal can be parallelized with care
   - Prim is difficult to parallelize effectively
