```python
import matplotlib.pyplot as plt
import numpy as np
filename1 = "BubbleSortRuntimes.txt"
filename2 = "MergeSortRuntimes.txt"
filename3 = "InsertionSortRuntimes.txt"
filename4 = "QuickSortRuntimes.txt"
filename5 = "SelectionSortRuntimes.txt"
x = np.arange(10000, 210000, 10000)
with open(filename1, "r") as file:
    bubbleArr = [float(line.strip()) for line in file]
with open(filename2, "r") as file:
    mergeArr = [float(line.strip()) for line in file]
with open(filename3, "r") as file:
    insertionArr = [float(line.strip()) for line in file]
with open(filename4, "r") as file:
    quickArr = [float(line.strip()) for line in file]
with open(filename5, "r") as file:
    selectionArr = [float(line.strip()) for line in file]

plt.figure(figsize = (10, 6))

plt.plot(x, bubbleArr, label = "Bubble Sort Runtime", marker = 'o', color="blue")
plt.plot(x, mergeArr,label = "Merge Sort Runtime", marker = 'o' ,color="green")
plt.plot(x, insertionArr,label = "Insertion Sort Runtime", marker = 'o' ,color="yellow")
plt.plot(x, quickArr, label = "Quick Sort Runtime", marker = 'o' ,color="purple")
plt.plot(x, selectionArr, label = "Selection Sort Runtime" ,marker = 'o' ,color="orange")
plt.title("Ramas Lab Exercise 2: Sorting Algorithms Analysis")
plt.xlabel("Size of Input")
plt.ylabel("Runtime")
plt.legend(loc = "upper left", title = "Legend")
plt.grid(True)
plt.show()
```

Algorithm Analysis Findings:

The graph above exhibits a visual guide on the execution time of sorting algorithms at incrementing intervals of input sizes. Starting from an input size of 10000, and incrementing its way up to 200000 with an interval of 10000, the graph with the highest elevation (blue) belongs to bubble sort, making it the slowest among the rest of the algorithms showcased in the graph with an asymptotic runtime of O(N^2). The second highest elevation belongs to selection sort (Orange), with an asymptotic runtime of O(N^2), despite having the same asmptotic runtime however, we can see that bubble sort is still significantly slower by comparison. Third slowest, and only slightly faster than selection sort is insertion. The remaining algorithms, quick sort and merge sort, run at a significantly faster rate relative to the running times of the previous ones. This is because, unlike the previous three, they have an asymptotic runtime of O(nlogn), making it much more efficient by a significant degree. In the graph, there is hardly any elevation at all with their nigh-instantaneous time of execution, even with input sizes that reached all the way up to 200000. In summary, the graph essentially showcases just how much more efficient sorting algorithms that run at O(nlogn) are with respect to the ones that run at O(N^2).
